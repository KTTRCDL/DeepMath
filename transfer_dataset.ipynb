```python
import pandas as pd

# read parquet file
train_all_df = pd.read_parquet('~/data/DeepMath-103k/train.parquet')
test_all_df = pd.read_parquet('~/data/DeepMath-103k/test.parquet')
```

```python
# get the full category of the type column of the train set and test set
train_all_df['topic'].unique(), test_all_df['type'].unique()
```

```python
Precalculus_data_df = train_all_df[train_all_df['topic'].str.contains('Mathematics -> Precalculus')]
Algebra_data_df = train_all_df[train_all_df['topic'].str.contains('Mathematics -> Algebra')]
Calculus_data_df = train_all_df[train_all_df['topic'].str.contains('Mathematics -> Calculus')]
DifferentialEquations_data_df = train_all_df[train_all_df['topic'].str.contains('Mathematics -> Differential Equations')]
NumberTheory_data_df = train_all_df[train_all_df['topic'].str.contains('Mathematics -> Number Theory')]
DiscreteMathematics_data_df = train_all_df[train_all_df['topic'].str.contains('Mathematics -> Discrete Mathematics')]
Geometry_data_df = train_all_df[train_all_df['topic'].str.contains('Mathematics -> Geometry')]
AppliedMathematics_data_df = train_all_df[train_all_df['topic'].str.contains('Mathematics -> Applied Mathematics')]
```

```python
Precalculus_data_df.to_parquet('~/data/DeepMath-103k-split/Precalculus.parquet')
Algebra_data_df.to_parquet('~/data/DeepMath-103k-split/Algebra.parquet')
Calculus_data_df.to_parquet('~/data/DeepMath-103k-split/Calculus.parquet')
DifferentialEquations_data_df.to_parquet('~/data/DeepMath-103k-split/DifferentialEquations.parquet')
NumberTheory_data_df.to_parquet('~/data/DeepMath-103k-split/NumberTheory.parquet')
DiscreteMathematics_data_df.to_parquet('~/data/DeepMath-103k-split/DiscreteMathematics.parquet')
Geometry_data_df.to_parquet('~/data/DeepMath-103k-split/Geometry.parquet')
AppliedMathematics_data_df.to_parquet('~/data/DeepMath-103k-split/AppliedMathematics.parquet')
```
